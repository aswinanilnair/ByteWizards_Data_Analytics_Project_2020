```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
from sklearn.model_selection import train_test_split
```

```python
cols = list(df.columns.values)
```

```python
cols = ['New Job Title', 'Job Position', 'Company Name', 'Headquarters', 'Rating', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'final']
```

```python
df = df[cols]
```

```python
df.head()
```

```python
y = df['final']
X = df.drop('final', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
y_train.head()
```

```python
tfidf_jobid = tfidf_vectorizer.fit_transform((y_train))
```

```python
print(tfidf_jobid)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
u = 657
t = [[u, y_test.iloc[0]]]
```

```python
temp_test = pd.DataFrame(t, columns=['ID', 'stuff'])
```

```python
temp_test.head()
```

```python
user_tfidf = tfidf_vectorizer.transform(temp_test['stuff'])
```

```python
output = list(map(lambda x: cosine_similarity(user_tfidf, x),tfidf_jobid))
```

```python
top = sorted(range(len(output)), key=lambda i: output[i], reverse=True)[:50]
```

```python
recommendation = pd.DataFrame(index=range(0,len(top)), columns = ['ApplicantID', 'JobID'])
count = 0
for i in top:
    try:
        print(list(X_test.index).index(i))
        recommendation.iloc[count]['ApplicantID'] = u
        recommendation.iloc[count]['JobID'] =  X_test.iloc[i]['New Job Title']
    except:
        print("Failed: ", i)

    count += 1
X_train.head()
```
